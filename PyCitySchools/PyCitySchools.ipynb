```python
# import dependencies
import pandas as pd
import numpy as np
import os 
import csv
```

```python
# file path to the inputs
filepath_schools = os.path.join("raw_data","schools_complete.csv")
filepath_students = os.path.join("raw_data","students_complete.csv")
```

```python
# read-in the inputs using pandas
df_schools = pd.read_csv(filepath_schools,sep=',')
df_students = pd.read_csv(filepath_students,sep=',')
```

```python
# print the head of the dataframes
display(df_schools.head())
display(df_students.head())
```

```python
# print the shape of the dataframes
display(df_schools.shape)
display(df_students.shape)
```

# ------------------------------------------------------------------------------------------

**District Summary**

* Create a high level snapshot (in table form) of the district's key metrics, including:
  * Total Schools
  * Total Students
  * Total Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#from school df
summary_district1 = pd.DataFrame({ "Total Schools" : df_schools['School ID'].nunique(),
"Total Students" : df_schools['size'].sum(),
"Total Budget" : df_schools['budget'].sum()},index=[0])
summary_district1.style.set_properties(**{'background-color': 'ash',
                           'color': 'blue',
                           'border-color': 'white'})
```

```python
# #from student df
summary_district2 = pd.DataFrame({"Average Math Score" : df_students['math_score'].mean(),
"Average Reading Score" : df_students['reading_score'].mean(),
"% Passing Math": (df_students['math_score'].apply(lambda x: x >= 75)).mean(),
"% Passing Reading": (df_students['reading_score'].apply(lambda x: x >= 75)).mean()},index=[0])
                                  
summary_district2["Overall Passing Rate"] = (summary_district2["% Passing Math"]+ \
                                            summary_district2["% Passing Reading"] )/2
    
summary_district2.style.set_properties(**{'background-color': 'ash',
                           'color': 'blue',
                           'border-color': 'white'})
```

```python
district_summary = pd.concat([summary_district1, summary_district2], axis=1)
district_summary.style.set_properties(**{'background-color': 'ash',
                           'color': 'blue',
                           'border-color': 'white'})
```

# ------------------------------------------------------------------------------------------

**School Summary**

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per School Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
summary_school1 = df_schools.groupby(["name","type"]).agg({"size":np.sum,"budget":np.sum}).reset_index()
summary_school1.columns = ["School Name","School Type","Total Students", "Total School Budget"]
summary_school1["Per School Budget"] = summary_school1["Total School Budget"]/summary_school1["Total Students"]
summary_school1 = summary_school1.set_index("School Name")
summary_school1
```

```python
summary_school2 = df_students.groupby(["school"]).agg({'math_score':np.mean,'reading_score':np.mean}).reset_index()
summary_school2 = summary_school2.set_index("school")
summary_school2
```

```python
def passrate(x):
    return (x['math_score'].apply(lambda x: x >= 75).mean(),
            x['reading_score'].apply(lambda x: x >= 75).mean())

summary_school3 = df_students.groupby('school').apply(passrate).reset_index()
summary_school3[['% Passing Math', '% Passing Reading']] = summary_school3[0].apply(pd.Series)
summary_school3 = summary_school3.set_index("school")
summary_school3 = summary_school3[['% Passing Math', '% Passing Reading']]
summary_school3
```

```python
summary_school = pd.concat([summary_school1,summary_school2,summary_school3],axis=1)
summary_school['Overall Passing Rate']= (summary_school['% Passing Math']+summary_school['% Passing Reading'])/2
summary_school.style.set_properties(**{'background-color': '#324376',
                           'color': 'white',
                           'border-color': 'white'})
```

# ------------------------------------------------------------------------------------------

**Scores by School Spending**

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
df_students_new = df_students.merge(df_schools,left_on='school',right_on='name')
df_students_new['Per Student Budget']=df_students_new['budget']/df_students_new['size']
```

```python
df_students_new.head()
```

```python
import matplotlib.pyplot as plt
df_students_new['Per Student Budget'].plot(kind='hist')
plt.show()
```

```python
perstudent_bins = [0,600,650,700]
perstudent_labels = ['<600', '600-650', '650+']
df_students_new['Per Student Budget Category'] = pd.cut(df_students_new['Per Student Budget'], perstudent_bins ,labels = perstudent_labels ,right=False)
```

```python
summary_scores1 = df_students_new.groupby(['Per Student Budget Category']).agg({'math_score':np.mean,'reading_score':np.mean})
summary_scores1
```

```python
def passrate(x):
    return (x['math_score'].apply(lambda x: x >= 75).mean(),
            x['reading_score'].apply(lambda x: x >= 75).mean())

summary_scores2 = df_students_new.groupby(['Per Student Budget Category']).apply(passrate).reset_index()
summary_scores2[['% Passing Math', '% Passing Reading']] = summary_scores2[0].apply(pd.Series)
summary_scores2 = summary_scores2.set_index("Per Student Budget Category")
summary_scores2 = summary_scores2[['% Passing Math', '% Passing Reading']]
summary_scores2
```

```python
summary_scores = pd.concat([summary_scores1,summary_scores2],axis=1)
summary_scores ['Overall Passing Rate']= (summary_scores['% Passing Math']+summary_scores['% Passing Reading'])/2
summary_scores.style.bar(color='#05668D')
```
